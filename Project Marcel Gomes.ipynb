```python
# Import library
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import pandas as pd

# Start driver Safari
driver = webdriver.Safari()

# URL site Motor1
url = "https://motor1.uol.com.br/news/"

# Data list
noticias = []

# Number of the page
numero_de_paginas = 10  # Example, 10 pages

for pagina in range(1, numero_de_paginas + 1):
    # Visit page
    driver.get(f"{url}?p={pagina}")
    
    # Waiting news page load
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, "item")))
    
    # Find all the news on the page
    noticias_na_pagina = driver.find_elements(By.CLASS_NAME, "item")

# Extract information from each news
    for item in noticias_na_pagina:
        titulo = item.find_element(By.TAG_NAME, "h3").text
        autor = item.find_element(By.CLASS_NAME, "name").text
        data = item.find_element(By.CLASS_NAME, "date").text
        # Try to find the comment count element, handling the exception if it doesn't exist
        try:
            comentarios = item.find_element(By.CLASS_NAME, "commentsNumber_num").text
        except NoSuchElementException:
            comentarios = "N/A"  # Or any other message you want
        
        noticias.append([titulo, autor, data, comentarios])

# Creat DataFrame with data news
data_news = pd.DataFrame(noticias, columns=["Título", "Autor", "Data", "Contagem de Comentários"])

# Show DataFrame
print(data_news)

# Close browser
driver.quit()
```

```python
from datetime import datetime, timedelta, date
import re

# Função para extrair datas
def extrair_data(texto):
    # Dicionário para mapear nomes de mês em português para inglês
    meses_pt_para_en = {
        'Janeiro': 'January',
        'Fevereiro': 'February',
        'Março': 'March',
        'Abril': 'April',
        'Maio': 'May',
        'Junho': 'June',
        'Julho': 'July',
        'Agosto': 'August',
        'Setembro': 'September',
        'Outubro': 'October',
        'Novembro': 'November',
        'Dezembro': 'December'
    }

    # Verifica se o texto possui o formato "Dia Mês Ano" em português
    padrao_data_pt = r'\d+\s+[A-Za-z]+\s+\d+'
    if re.match(padrao_data_pt, texto):
        # Mapeia nomes de meses em português para inglês
        for mes_pt, mes_en in meses_pt_para_en.items():
            texto = texto.replace(mes_pt, mes_en)
        
        data_obj = datetime.strptime(texto, '%d %B %Y')
        return data_obj.strftime('%b-%d-%Y')
    else:
        # Verifica se o texto possui o formato "Xh atrás", "Xm atrás" ou "Xs atrás"
        padrao_horas_minutos_segundos = r'(\d+)\s*([hms])\s+atrás'
        match = re.match(padrao_horas_minutos_segundos, texto)
        if match:
            quantidade = int(match.group(1))
            unidade = match.group(2)
            agora = datetime.now()
            if unidade == 'h':
                nova_data = agora - timedelta(hours=quantidade)
            elif unidade == 'm':
                nova_data = agora - timedelta(minutes=quantidade)
            else:
                nova_data = agora - timedelta(seconds=quantidade)
            return nova_data.strftime('%b-%d-%Y')
        else:
            # Tente corresponder ao formato "Dia Mês Ano" abreviado
            padrao_data_abreviado = r'\d+\s+[A-Za-z]+\s+\d+'
            if re.match(padrao_data_abreviado, texto):
                # Remova possíveis espaços extras
                texto = re.sub(r'\s+', ' ', texto)
                data_obj = datetime.strptime(texto, '%d %b %Y')
                return data_obj.strftime('%b-%d-%Y')
            else:
                # Se não corresponder a nenhum dos formatos acima, defina como "Data Não Disponível"
                return "Data Não Disponível"

# Aplique a função de extração à coluna 'Data'
data_news['Data Formatada'] = data_news['Data'].apply(extrair_data)

# Exiba o DataFrame atualizado
print(data_news)
```

```python
from textblob import TextBlob

# Função para determinar o sentimento com base no TextBlob
def determinar_sentimento(texto):
    analysis = TextBlob(texto)
    polaridade = analysis.sentiment.polarity
    if polaridade > 0:
        return 1  # Sentimento positivo
    elif polaridade < 0:
        return -1  # Sentimento negativo
    else:
        return 0  # Sentimento neutro

# Ler as notícias do arquivo CSV (substitua "noticias.csv" pelo caminho correto do seu arquivo)
import pandas as pd

data_news = pd.read_csv("noticias.csv")

# Aplicar a função de determinar_sentimento às notícias
data_news['Sentimento'] = data_news['Título'].apply(determinar_sentimento)

# Agora, o DataFrame "data_news" tem uma coluna "Sentimento" com valores -1 (negativo), 0 (neutro) ou 1 (positivo) para cada notícia.

# Exemplo: imprimir a primeira notícia e seu sentimento
print("Texto da primeira notícia:")
print(data_news['Título'][0])
print("Sentimento da primeira notícia:")
print(data_news['Sentimento'][0])
```
